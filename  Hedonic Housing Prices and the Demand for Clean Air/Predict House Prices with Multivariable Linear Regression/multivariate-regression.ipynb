```python
# VIF-> Variance Inflation Factor
# Variance inflation factors (VIF) measure how much the variance of the estimated regression coefficients are inflated as compared to when the predictor variables are not linearly related.
```

# Testing for multicollineariaty
#### $$ TAX= \alpha _0 + \alpha _1 RM + \alpha _2 NOX + ... + \alpha _{12}LSTAT NOX $$
#### $$ VIF _{TAX} = \frac{1}{(1-R _{TAX} ^2)} $$

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# type(x_incl_cons)
variance_inflation_factor(exog=x_incl_cons.values , exog_idx=1)
```

```python
print(len(x_incl_cons.columns))
# x_incl_cons.shape[0]
x_incl_cons.shape[1]
```

```python
for i in range(x_incl_cons.shape[1]):
    print(variance_inflation_factor(exog=x_incl_cons.values, exog_idx = i))
print('!!!Done')
```

```python
# for storing
vif = []
for i in range(x_incl_cons.shape[1]):
    vif.append(variance_inflation_factor(exog=x_incl_cons.values, exog_idx = i))
print(vif)
```

```python
vif = [(variance_inflation_factor(exog=x_incl_cons.values, exog_idx = i)) for i in range(x_incl_cons.shape[1]) ]

pd.DataFrame({'coef_name':x_incl_cons.columns, 'vif': np.around(vif, 2) })
# print(vif)
```

## Model Simplification and BIC

```python
# original model with log prices and all features 

x_incl_cons = sm.add_constant(X_train)
model = sm.OLS(Y_train, x_incl_cons)  #ordinary least square
results = model.fit()  #regresssion results


org_coef = pd.DataFrame({'coef': results.params , 'pvalue': round(results.pvalues,3)})

# print BIC and r-square

print('BIC is',results.bic)
print('R-square is',results.rsquared)
```

```python
# original model excluding INDUS
x_incl_cons = sm.add_constant(X_train)
x_incl_cons = x_incl_cons.drop(['INDUS'], axis=1)

model = sm.OLS(Y_train, x_incl_cons)  #ordinary least square
results = model.fit()  #regresssion results


coef_minus_indus = pd.DataFrame({'coef': results.params , 'pvalue': round(results.pvalues,3)})

# print BIC and r-square
print('BIC is',results.bic)
print('R-square is',results.rsquared)
```

# Residuals and Residual plot

```python
prices = np.log(data['PRICE'])
features = data.drop(['PRICE', 'INDUS', 'AGE'], axis=1)
X_train,X_test, Y_train, Y_test = train_test_split(features, prices, test_size = .2, random_state = 10)


# Using statemodels
x_incl_cons = sm.add_constant(X_train)
model = sm.OLS(Y_train, x_incl_cons)  #ordinary least square
results = model.fit()  #regresssion results

# residuals = Y_train - results.fittedvalues
# residuals

# results.resid
# Greph of Actual and predicted result
corr = round(Y_train.corr(results.fittedvalues), 2)
# corr

plt.figure(figsize = (15,8))
plt.xlabel('Actual log-prices $Y _i$', fontsize=16)
plt.ylabel('Predicted log prices $ \hat y_i$', fontsize=16)
plt.title(f'Actual-vs-Predicted log prices: $Y _i$ vs $ \hat y_i$(Corr {corr})', fontsize=18)
plt.plot(Y_train, Y_train)

plt.scatter(x=Y_train, y = results.fittedvalues, c='navy', alpha=.6, )
plt.show()

plt.figure(figsize = (15,8))
plt.xlabel('Actual prices(in thousands) $Y _i$', fontsize=16)
plt.ylabel('Predicted prices(in thousands) $ \hat y_i$', fontsize=16)
plt.title(f'Actual-vs-Predicted prices: $Y _i$ vs $ \hat y_i$(Corr{corr})', fontsize=18 )
plt.plot(np.e**Y_train, np.e**Y_train, color='red')
plt.scatter(x=np.e**Y_train, y = np.e**results.fittedvalues,  alpha=.6,c='green' )
plt.show()

plt.figure(figsize = (15,8))
plt.xlabel('Fitted-values', fontsize=16)
plt.ylabel('Residuals', fontsize=16)
plt.title('Residuals vs Fitted-val', fontsize=18 )
plt.scatter(x=results.fittedvalues, y = results.resid,  alpha=.6,c='purple' )
plt.show()

# mean square error
reduced_log_mse = round(results.mse_resid, 3)
reduced_log_rsquared = round(results.rsquared, 3)
```

```python
# Distribution of residuals
# print(results.resid.describe())
resid_skew = round(results.resid.skew(), 3)
resid_mean = round(results.resid.mean(), 3)

plt.figure(figsize = (12,7))

sns.distplot(results.resid, color='pink')
plt.title(f"(Log Price model: results_skew({resid_skew}) mean({resid_mean})")
plt.plot()

# mean square error
```

```python
# for our estimates of a house as $30000,
# Can we calculate the upper and lower bound of house for 95% prediction model using reduced-log model

estimates_price  = 30 #(in thousand)
print('1) SD in log price is ',np.sqrt(reduced_log_mse))
print('2) SD in log price is ',2*np.sqrt(reduced_log_mse))

upperbound = np.log(estimates_price) + 2*np.sqrt(reduced_log_mse)
print('The upper bound in log price with 95% prediction interval: ',upperbound)
print('The upper bound in original price with 95% prediction interval: $ ',np.e**upperbound*1000)

lowerbound = np.log(estimates_price) + 2*np.sqrt(reduced_log_mse)
print('The lowerbound  in log price with 95% prediction interval: ',lowerbound)
print('The lowerbound  in original price with 95% prediction interval: $ ',np.e**lowerbound*1000)
```
